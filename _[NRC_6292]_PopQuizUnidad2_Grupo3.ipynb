# Medidas de Frecuencia

```python
'''Función para calcular frecuentcia'''
def rel_freq(x):
    freqs = [(value, x.count(value) / len(x)) for value in set(x)] 
    return freqs
```

```python
'''Verificar la frecuencia de las variables de Pais'''
df['Pais'].value_counts()
```
